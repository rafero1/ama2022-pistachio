```python
import numpy as np
import math
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import make_scorer, accuracy_score, f1_score, precision_score, recall_score, roc_curve, auc, roc_auc_score, confusion_matrix, plot_confusion_matrix, ConfusionMatrixDisplay, RocCurveDisplay
```

```python
labels = ['Kirmizi_Pistachio', 'Siirt_Pistachio']
confusion_matrix_16_features = pd.read_csv('confusion_matrix_16_features.csv', index_col=0)
roc_curve_16_features = pd.read_csv('roc_curve_16_features.csv', index_col=0)
confusion_matrix_28_features = pd.read_csv('confusion_matrix_28_features.csv', index_col=0)
roc_curve_28_features = pd.read_csv('roc_curve_28_features.csv', index_col=0)
```

```python
confusion_matrix_16_features.to_numpy()
```

```python
sns.reset_orig()

fig_confusion, ax_confusion = plt.subplots(1, 2, figsize=(14, 4))

cm16 = confusion_matrix_16_features.to_numpy()
display16 = ConfusionMatrixDisplay(confusion_matrix=cm16, display_labels=labels)
display16.plot(ax=ax_confusion[0])
ax_confusion[0].set_title('Dataset 1 (16 features)')

cm28 = confusion_matrix_28_features.to_numpy()
display28 = ConfusionMatrixDisplay(confusion_matrix=cm28, display_labels=labels)
display28.plot(ax=ax_confusion[1])
ax_confusion[1].set_title('Dataset 2 (28 features)')

plt.savefig('knn-con-matrixes.png')
plt.show()
```

```python
sns.set_theme()

fig_roc_curve, ax_roc_curve = plt.subplots(1, 2, figsize=(12, 4))

fpr, tpr, roc_auc = roc_curve_16_features.to_numpy()
display16 = RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc[0], estimator_name='KNN')
display16.plot(ax=ax_roc_curve[0])
ax_roc_curve[0].set_title('Dataset 1 (16 features)')

fpr, tpr, roc_auc = roc_curve_28_features.to_numpy()
display28 = RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc[0], estimator_name='KNN')
display28.plot(ax=ax_roc_curve[1])
ax_roc_curve[1].set_title('Dataset 2 (28 features)')

plt.savefig('knn-roc-curves.png')
plt.show()
```
