## Import

```python
import pandas as pd
import os
import numpy as np
from sklearn.model_selection import train_test_split
import shutil
```

## Create Folders

```python
os.mkdir("Pistachio_Image_Dataset_Test")
os.mkdir("Pistachio_Image_Dataset_Test/Kirmizi_Pistachio")
os.mkdir("Pistachio_Image_Dataset_Test/Siirt_Pistachio")

os.mkdir("Pistachio_Image_Dataset_Val")
os.mkdir("Pistachio_Image_Dataset_Val/Kirmizi_Pistachio")
os.mkdir("Pistachio_Image_Dataset_Val/Siirt_Pistachio")
```

## Split Test

```python
class1 = sorted(os.listdir("Pistachio_Image_Dataset_Train/Kirmizi_Pistachio"))
class2 = sorted(os.listdir("Pistachio_Image_Dataset_Train/Siirt_Pistachio"))
```

```python
label1 = ["Kirmizi" for a in range(len(class1))]
label2 = ["Siirt" for a in range(len(class2))]
```

```python
d = {'Images': class1+class2, 'Labels': label1+label2}
```

```python
df = pd.DataFrame(data=d)
```

```python
df
```

```python
X = df.iloc[:,0:1]
y = df.iloc[:,-1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=42,stratify=y)
```

```python
for a in X_test["Images"]:
    path = os.path.join("Pistachio_Image_Dataset_Test","Kirmizi_Pistachio")
    path1 = os.path.join("Pistachio_Image_Dataset_Test","Siirt_Pistachio")
    path_ori = os.path.join("Pistachio_Image_Dataset_Train","Kirmizi_Pistachio")
    path_ori1 = os.path.join("Pistachio_Image_Dataset_Train","Siirt_Pistachio")
    
    pistachio_c = a.split()[0]
    name = a.split()
    print (name)
    
    if pistachio_c == 'kirmizi':
        shutil.move(path_ori + "/"+name[0]+" "+name[1],path+ "/"+name[0]+" "+name[1])
    else:
        shutil.move(path_ori1 + "/"+name[0]+" "+name[1],path1+ "/"+name[0]+" "+name[1])
```
